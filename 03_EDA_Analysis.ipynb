## **2. Visualize the Distribution of Key Variables**
***
To get a better sense of our data, we'll plot the distribution of several important variables.
> - **Histograms** are used for numeric columns (like profit, sales, quantity) to show their spread and skew.
> - **Bar plots** (count plots) are used for categorical fields (like Delivery Status, Category Name, Shipping Mode) to see the most common values.
***
<hr style="border: 1px solid red;">

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Modern, clean look
sns.set_theme(style="whitegrid", palette="pastel")
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

<hr style="border: 1px solid black;">

```python
num_cols = ["Benefit per order", "Sales per customer", "Order Item Quantity"]
custom_colors = ["#64b5f6", "#ffb74d", "#81c784"]  # blue, orange, green

for col, c in zip(num_cols, custom_colors):
    plt.figure(figsize=(10,6))
    sns.histplot(df[col], bins=40, kde=True, color=c, edgecolor="black", alpha=0.9)
    plt.title(f"Distribution of {col}", fontweight="bold", pad=15)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.grid(axis='y', linestyle="--", alpha=0.3)
    sns.despine()
    plt.tight_layout()
    plt.show()
```

<hr/>
<div class="alert alert-success alertsuccess" style="margin-top: 20px">
    <h3><b>1. Benefit per order</b></h3>
    <ul>
        <li><b>Most values are close to zero</b> (small profit or small loss).</li>
        <li>A long tail to the left (some very large negative values) and a smaller tail to the right (some very large profits), which means there are extreme negative outliers but most orders are only mildly profitable or unprofitable.</li>
        <li><b>Interpretation:</b> Most orders make a modest profit; a few orders may lose a lot or make an unusually high gain. Outliers should be explored further.</li>
    </ul>
    <hr/>
    <h3><b>2. Sales per customer</b></h3>
    <ul>
        <li><b>Most customers have sales under $400</b>, with the majority clustered well below that.</li>
        <li>The distribution is heavily <b>right-skewed</b>: a small number of customers spend much more than average.</li>
        <li><b>Interpretation:</b> Most customers make small purchases, but there are a few high-value customers who buy a lot. This is typical in sales data.</li>
    </ul>
    <hr/>
    <h3><b>3. Order Item Quantity</b></h3>
    <ul>
        <li>The vast majority of orders are for <b>1 item</b>.</li>
        <li>There are <b>distinct spikes at whole numbers</b>: 2, 3, 4, and 5 items.</li>
        <li>Almost no orders for fractional quantities (as expected for discrete products).</li>
        <li><b>Interpretation:</b> Most purchases are for single items, but a significant number of orders are for small bulk quantities (2-5). This sort of pattern is common in retail and supply chain transactions.</li>
    </ul>
    <hr/>
    <b>Summary:</b><br/>
    The sales and profit columns show a lot of orders with modest values, but a few very large (positive or negative) outliers. Most orders are for one item, but bulk buying is clearly visible in the data.
</div>
<hr/>
<hr style="border: 1px solid black;">

## **3.  Explore Relationships Between Key Variables**
***
Now, we’ll look for patterns and connections between different columns, starting with correlations and moving on to comparisons across categories.
***
<hr style="border: 1px solid red;">

#### **Correlation Table**
<hr style="border: 1px solid black;">

```python
plt.figure(figsize=(22, 18))
selected_cols = ["Benefit per order", "Sales per customer", "Order Item Quantity",
                 "Order Item Total", "Sales", "Order Profit Per Order",
                 "Order Item Product Price", "Order Item Discount",
                 "Order Item Profit Ratio", "Product Price",
                 "Days for shipping (real)", "Days for shipment (scheduled)"]
corr = df[selected_cols].corr()

sns.set_theme(style="white")
ax = sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    annot_kws={"size": 18, "weight": "bold"},
    cmap="icefire",
    linewidths=1.0,
    square=True,
    cbar_kws={"shrink": .75, "label": "Correlation"},
    vmin=-1, vmax=1
)

plt.title("Correlation Heatmap of Selected Numeric Features", fontsize=26, fontweight="bold", pad=30)
plt.xticks(rotation=30, ha='right', fontsize=15)
plt.yticks(fontsize=15)
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

<hr/>
<div class="alert alert-success alertsuccess" style="margin-top: 20px">
    <h3><b>Interpretation of Correlation Heatmap</b></h3>
    <ul>
        <li><b>Sales-related variables</b> (like Sales per customer, Order Item Total, and Sales) are extremely closely linked—they all increase together.</li>
        <li><b>Profit metrics</b> (Benefit per order and Order Profit Per Order) are also strongly related, meaning higher profits usually pair with higher benefit per order.</li>
        <li><b>Order Item Quantity and Product Price</b> have a moderate negative correlation: customers buy more units when items are cheaper.</li>
        <li><b>Shipping days</b> (real and scheduled) are only moderately related to each other and have little to no correlation with sales, profit, or quantity—delivery timing is mostly independent of sales performance.</li>
    </ul>
</div>
<hr/>
<hr style="border: 1px solid black;">


#### **Scatter Plot**
<hr style="border: 1px solid black;">

```python
# 1. Sales per customer vs. Benefit per order
plt.figure(figsize=(9,7))
sns.scatterplot(
    data=df,
    x="Sales per customer",
    y="Benefit per order",
    s=30,
    color="#1976d2",
    alpha=0.28,
    edgecolor='none'
)
plt.title("Sales per Customer vs. Benefit per Order", fontsize=18, weight="bold", pad=15)
plt.xlabel("Sales per Customer", fontsize=14)
plt.ylabel("Benefit per Order", fontsize=14)
plt.grid(alpha=0.18)
sns.despine()
plt.tight_layout()
plt.show()
```

<hr/>
<div class="alert alert-success alertsuccess" style="margin-top: 20px">
    <h3><b>1. Sales per Customer vs. Benefit per Order</b></h3>
    <ul>
        <li><b>Pattern:</b> Most points cluster at low-to-moderate sales per customer, with benefits (profit/loss per order) hovering near zero, but there are several outliers for both high profits and steep losses.</li>
        <li><b>Insight:</b> Generally, higher sales amount per customer does not guarantee high profit per order; there is a wide spread, including significant losses even at higher sales levels. This suggests factors like discounts, cost structure, or returns significantly impact order benefit.</li>
    </ul>
</div>
<hr/>
<hr style="border: 1px solid black;">

<hr/>
